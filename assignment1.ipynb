```python
import pandas as pd

data=pd.read_csv("Ecommerce Purchases.csv.csv")

language_counts = data['Language'].value_counts()

french_count = language_counts.get('fr', 0)
print("Number of people with French as their language:",french_count)
# Plot the bar graph
sns.countplot(x='Language', data=data)
plt.pyplot.title('Number of People Speaking Each Language')
plt.pyplot.xlabel('Language')
plt.pyplot.ylabel('Number of People')
plt.pyplot.show()
```

```python
import pandas as pd

data=pd.read_csv("Ecommerce Purchases.csv.csv")

engineer_data = data[data['Job'].str.contains('engineer', case=False, na=False)]
print(engineer_data)
#Count total engineers
engineer_count = data[data['Job'].str.contains('Engineer', case=False)].shape[0]
print(f'Total number of people with a job title containing "Engineer": {engineer_count}')
sns.countplot(x='Job', data=data[data['Job'].str.contains('Engineer', case=False)])
plt.pyplot.xticks(rotation=90)
plt.pyplot.title('Number of People with a Job Title Containing "Engineer"')
plt.pyplot.xlabel('Job Title')
plt.pyplot.ylabel('Number of People')
plt.pyplot.show()
```
