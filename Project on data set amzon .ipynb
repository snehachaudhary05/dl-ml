# find email of the person with the following IP address: 132.207.160.22

```python
import pandas as pd

df = pd.read_csv("Ecommerce Purchases.csv.csv")
ip_address_to_find = '132.207.160.22'
filtered_df = df[df['IP Address'] == ip_address_to_find]

if not filtered_df.empty:
    emails = filtered_df['Email'].tolist()  # Assuming the column name is 'Email'
    print("Email(s) for IP address", ip_address_to_find, ":", emails)
else:
    print("No matching records found for IP address", ip_address_to_find)
```

# Find Email of the person with the following Credit Card number:4105595335494659

```python
import pandas as pd

df=pd.read_csv("Ecommerce Purchases.csv.csv")

credit_card_number_to_find = '4105595335494659'

# Filter the DataFrame for the specified credit card number
filtered_df = df[df['Credit Card'] == credit_card_number_to_find]

# Display the email(s) corresponding to the specified credit card number
if not filtered_df.empty:
    emails = filtered_df['Email'].tolist()
    print("Email(s) for Credit Card Number", credit_card_number_to_find, ":", emails)
else:
    print("No matching records found for Credit Card Number", credit_card_number_to_find)
```

# how Many people Purchase During The Am and How Many People Purchase During Pm?

```python
import pandas as pd

df=pd.read_csv("Ecommerce Purchases.csv.csv")

am_purchases = df[df['AM or PM'] == 'AM']
pm_purchases = df[df['AM or PM'] == 'PM']


num_am_purchases = len(am_purchases)
num_pm_purchases = len(pm_purchases)


print("Number of people who made purchases during AM:", num_am_purchases)
print("Number of people who made purchases during PM:", num_pm_purchases)
```

# Top 5 Most Popular EMail providers(eg Gmail.com, Yahoo.com,etc.)

```python
import pandas as pd

df=pd.read_csv("Ecommerce Purchases.csv.csv")

df['Email Domain'] = df['Email'].str.extract(r'@(\w+\.\w+)')

email_counts = df['Email Domain'].value_counts()

top_5_email_providers = email_counts.head(5)
print("Top 5 Most Popular Email Providers:")
print(top_5_email_providers)
```
